```python
import os
import urllib.parse
import certifi
import pandas as pd
import pymongo
from dotenv import load_dotenv
```

```python
# 1. Load environment variables
load_dotenv()

db_name = os.getenv("DB_NAME")
collection_name = os.getenv("COLLECTION_NAME")
password = os.getenv("DB_PASSWORD")
username = "antaladhaval02ad_db_user"
```

```python
# 2. Read and convert CSV data
df = pd.read_csv("data.csv")
data = df.to_dict(orient="records")
```

```python
# 3. Safely encode credentials to avoid connection breaks
safe_username = urllib.parse.quote_plus(username)
safe_password = urllib.parse.quote_plus(password)
```

```python
# 4. Construct the authenticated connection URL
connection_url = f"mongodb+srv://{safe_username}:{safe_password}@cluster0.95ablet.mongodb.net/?appName=Cluster0"
```

```python
# 5. Connect to MongoDB and target your collection
client = pymongo.MongoClient(connection_url, tlsCAFile=certifi.where())
data_base = client[db_name]
collection = data_base[collection_name]  # Fixed: Defined the collection variable
```

```python
# 6. Upload data to MongoDB
if data:  # Ensures the CSV isn't empty
    rec = collection.insert_many(data)
    print(f"Successfully inserted {len(rec.inserted_ids)} records.")
```

```python
# 7. Load back data from MongoDB into a DataFrame
df_from_db = pd.DataFrame(list(collection.find()))
print(df_from_db.head(2))
```
